```python
#importing pandas library
import pandas as pd
```

```python
#Data collection
Dataset=pd.read_csv('insurance_pre.csv')
```

```python
Dataset
```

```python
Dataset=pd.get_dummies(Dataset,dtype=int,drop_first=True)
Dataset
```

```python
Independent=Dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
Independent
```

```python
Dependent=Dataset[['charges']]
Dependent
```

```python
#Standardisation

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
Independent=sc.fit_transform(Independent)
```

```python
#GridSeachCV

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
param_grid = {'fit_intercept':[False,True],'copy_X':[False,True],'n_jobs':[None],'positive':[True,False]}
grid = GridSearchCV(LinearRegression(), param_grid, refit = True, verbose = 3,n_jobs=-1)
grid.fit(Independent,Dependent)
```

```python
#Viewing the results in the dictionary as well as table format

re=grid.cv_results_
table=pd.DataFrame.from_dict(re)
table
```

```python
Linear_prediction = grid.predict(Independent)
```

```python
#R_Score

from sklearn.metrics import r2_score
r_score=r2_score(Dependent,Linear_prediction)
print(' R_score for best parameter{}:'.format(grid.best_params_))
```
