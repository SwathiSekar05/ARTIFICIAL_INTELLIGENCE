```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Dataset=pd.read_csv('insurance_pre.csv')
```

```python
Dataset
```

```python
Dataset=pd.get_dummies(Dataset,dtype=int,drop_first=True)
Dataset
```

```python
Dataset.columns
```

```python
Independent=Dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
Independent
```

```python
Dependent=Dataset[[ 'charges']]
Dependent
```

```python
#standardisation

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
Independent=sc.fit_transform(Independent)
```

```python
#GridSearchCV

from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
param_grid = {'criterion':['mse','mae','friedman_mse'],'max_features': [None,'sqrt','log2'],'splitter':['best','random']}
grid = GridSearchCV(DecisionTreeRegressor(), param_grid, refit = True, verbose= 3,n_jobs=-1)

# fitting the model for grid search
grid.fit(Independent,Dependent)
```

```python
re=grid.cv_results_
re
```

```python
Table=pd.DataFrame. from_dict(re)
Table
```

```python
grid_predictions = grid.predict(Independent)
```

```python
print(' R_score for best parameter{}:'.format(grid.best_params_))
```

```python
from sklearn.metrics import r2_score
R_score=r2_score(Dependent,grid_predictions)
R_score
```

```python
#Saving the Model

import pickle
filename="finalized_model_Decision_Tree.sav"
pickle.dump(grid_predictions,open(filename,'wb'))
```
