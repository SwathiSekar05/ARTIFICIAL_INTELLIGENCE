```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Dataset=pd.read_csv('insurance_pre.csv')
```

```python
Dataset
```

```python
Dataset=pd.get_dummies(Dataset,dtype=int,drop_first=True)
Dataset
```

```python
Independent=Dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
Independent
```

```python
Dependent=Dataset[['charges']]
Dependent
```

```python
#Dataset Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(Independent, Dependent, test_size=0.30,random_state=0)
```

```python
#standardisation
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
#GridSearchCCV

from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = {'kernel':['rbf','poly','sigmoid','linear'],'C':[10,100,1000,2000,3000],'gamma':['auto','scale']}

grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3,n_jobs=-1)

#Substituting the values,fitting the model.
grid.fit(X_train,y_train)
```

```python
#If we print and see...It will show it in dictionary format(Poor readability)
re=grid.cv_results_
re
```

```python
grid_predictions=grid.predict(X_test)
```

```python
#to view it in table format
Table=pd.DataFrame. from_dict(re)
Table
```

```python
#R_Score:
from sklearn.metrics import r2_score
r_score=r2_score(y_test,grid_predictions)
print(' R_score for best parameter{}:'.format(grid.best_params_))
```
