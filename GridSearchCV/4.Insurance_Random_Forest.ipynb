```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
Dataset=pd.read_csv('insurance_pre.csv')
```

```python
Dataset
```

```python
Dataset=pd.get_dummies(Dataset,dtype=int,drop_first=True)
Dataset
```

```python
Independent=Dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
Independent
```

```python
Dependent=Dataset[[ 'charges']]
Dependent
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
param_grid = {'criterion':['squarred_error','absolute_error','poisson','friedman_mse'],'max_features': [None,'sqrt','log2'],'n_estimators':[50,100]}

grid = GridSearchCV(RandomForestRegressor(), param_grid, refit = True, verbose= 3,n_jobs=-1)


grid.fit(Independent,Dependent)
```

```python
grid_predictions = grid.predict(Independent)
```

```python
re=grid.cv_results_
table=pd.DataFrame.from_dict(re)
table
```

```python
from sklearn.metrics import r2_score
R_score=r2_score(Dependent,grid_predictions)
R_score
```

```python
print("The R_score value for best parameter {}:".format(grid.best_params_))
```
