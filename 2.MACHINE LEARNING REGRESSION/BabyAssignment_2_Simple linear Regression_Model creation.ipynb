```python
#Here we are importing the pandas library in which it supports complete dataset.
#when we import this panda library here it will automatically get saved to the memory of this file.
import pandas as pd
```

```python
#1.Data collection
```

```python
#In this case,we are assigning the values which is present in the Salary_Data.csv file to the Dataset (Variable) 
Dataset=pd.read_csv("Salary_Data.csv")
```

```python
Dataset
```

```python
#2.Input/Output split
```

```python
#For assigning the specific column from the tabulation to the separate variable,select the respective column from variable  where it is assigned.

#Here we are assigning the Input variable column(YearsExperience) from the Dataset variable to Independent Variable.

Independent=Dataset[["YearsExperience"]]
```

```python
Independent
```

```python
#Here we are assigning the Output variable column(Salary) from the Dataset variable where to Dependent Variable.

Dependent=Dataset[["Salary"]]
```

```python
Dependent
```

```python
#3.Training & Test set split
```

```python
#here we are spliting the dataset into training and test set. For that we are using sklearn library.
#sklearn library supports for full Machine learning.
#In this case,from sklearn library we are calling required class and function which we need(model_selection,train_test_split)
# Then we are assigning respective variables(X_train,X_test,y_train,y_test=train_test_split) and parameters(Independent,Dependent,test_size=0.30,random_state=0)

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(Independent,Dependent,test_size=0.30,random_state=0)
```

```python
#4.Model Creation
```

```python
#By using Linear regression algorithm which we have learnt theoretically, here we are doing the respective coding.
#This is the exact phase where the learning is happening
#In Fit function we are substiuting the input X_train,y_train whereas after execution, weight & bias are calculated and we will get the respective model

from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
#Here we are finding the value of the weight(Slope) by using the function coef_ which is already calculated
weight=regressor.coef_
weight
```

```python
#Here we are finding the value of the bias(origin/Intercept)  by using the function intercept_ which is already calculated
bias=regressor.intercept_
bias
```

```python
# Evaluation metrics
```

```python
#prediction is the application function.By using this fn,we are giving the parameter(X_test) and assigning to the variable y_prediction
#So it will calculate the input values given in the test set(X_test)by the use of model we created.
y_prediction=regressor.predict(X_test)
```

```python
y_prediction
```

```python
#Here, by using r2_score we are analyzing how well the model is fitted.
#Inside the r2_score fn we are giving the parameter(y_test,y_prediction)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_prediction)
```

```python
r_score
```

```python
#By using pickle we are giving the respective filename and saving the model.
import pickle
Filename="Finalized_Simple_Linear_Model.sav"
```

```python
#Here we are implementing the model into the file we created "Finalized_Simple_Linear_Model.sav"
#wb-write binary

pickle.dump(regressor,open(Filename,'wb'))
```

```python
#From the saved model we are loading and checking it for the further use. Further use is the case for Phase 2(Deployment)
#rb-read binary
Loaded_Model=pickle.load(open("Finalized_Simple_Linear_Model.sav",'rb'))
Result=Loaded_Model.predict([[13]])
```
