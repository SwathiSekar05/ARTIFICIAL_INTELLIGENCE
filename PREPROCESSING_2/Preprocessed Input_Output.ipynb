```python
import pandas as pd
```

```python
Dataset=pd.read_csv('insurance_pre.csv')
```

```python
Dataset
```

```python
Dataset=pd.get_dummies(Dataset,dtype=int,drop_first=True)
Dataset
```

```python
Independent=Dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
Independent
```

```python
Dependent=Dataset[[ 'charges']]
Dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(Independent,Dependent,test_size=0.30,random_state=0)
```

```python
#standardisation
#Preprocessing both Input and Output

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)

scy=StandardScaler()
Y_train=scy.fit_transform(Y_train)
Y_test=scy.transform(Y_test)
```

```python
#GridSearchCV

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
param_grid = {'criterion':['squarred_error','absolute_error','poisson','friedman_mse'],'max_features': [None,'sqrt','log2'],'n_estimators':[50,100]}
grid = GridSearchCV(RandomForestRegressor(), param_grid, refit = True, verbose= 3,n_jobs=-1)

#Fitting the model
grid.fit(X_train,Y_train)
```

```python
re=grid.cv_results_
Table=pd.DataFrame. from_dict(re)
Table
```

```python
print(' R_score for best parameter{}:'.format(grid.best_params_))
```

```python
grid_predictions = grid.predict(X_test)
```

```python
#R_Score:

from sklearn.metrics import r2_score
R_score=r2_score(Y_test,grid_predictions)
R_score
```

```python
import pickle
filename="finalized_model_Random_Forest.sav"
pickle.dump(grid,open(filename,'wb'))
```
