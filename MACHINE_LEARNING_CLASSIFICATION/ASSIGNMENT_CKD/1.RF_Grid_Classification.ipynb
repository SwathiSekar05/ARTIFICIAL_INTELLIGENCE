```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Loading Dataset
dataset=pd.read_csv("CKD.csv")
dataset
```

```python
#checking counts-Rows&columns
dataset.shape
```

```python
#get dummies
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
dataset["classification_yes"].value_counts()
```

```python
dataset.columns
```

```python
#spliting Independent and Dependent
Independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv','wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes','appet_yes', 'pe_yes', 'ane_yes']]
Independent
```

```python
Dependent=dataset[["classification_yes"]]
Dependent
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Independent,Dependent, test_size = 1/3, random_state = 0)
```

```python
#StandardScaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_= sc.fit_transform(X_train)
X_test_= sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
param_grid = {'criterion':['gini','entropy','log_loss'],'max_features': [None,'sqrt','log2'],'n_estimators':[10,100]} 
grid = GridSearchCV(RandomForestClassifier(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1') 
   
# fitting the model for grid search 
grid.fit(X_train, y_train)

#we have lot scoring metrics....we can also us that one
#https://scikit-learn.org/stable/modules/model_evaluation.html#classification-metrics
        
```

```python
grid_predictions = grid.predict(X_test) 
```

```python
re=grid.cv_results_
table=pd.DataFrame.from_dict(re)
table
```

```python
print(grid.best_params_)
```

```python
#evaluation metrics(confusion_matrix)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(y_test,grid_predictions)
```

```python
print(clf_report)
```

```python
#f1 score prediction

from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_predictions,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

```python
#average{‘micro’, ‘macro’, ‘samples’, ‘weighted’, ‘binary’} or None, default=’binary’
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(X_test)[:,1])
```
